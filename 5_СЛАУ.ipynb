```python
import numpy as np
import math
```

```python
for n in (4, 8, 16, 32, 64):
  sum = 0
  A = np.arange(n*n, dtype=float).reshape(n, n)
  for i in range (0, n):
    for j in range (0, n):
      A[i][j] = 1 / (i + 1 + j + 1 - 1)
  B = np.arange(n, dtype = float).reshape(n, 1)
  for j in range (0, n):
    for i in range (0, n):
      sum = sum + A[i][j]
    B[j][0] = sum
    sum = 0
  Btransp = np.arange(n, dtype = float)
  for i in range (0, n):
    Btransp[i] = B[i][0]

  def gauss_metod(a,b):
      n = len(b)
     # Elimination phase
      for k in range(0,n-1):
          for i in range(k+1,n):
              if a[i,k] != 0.0:
                  #if not null define λ
                  lam = a [i,k]/a[k,k]
                  #we calculate the new row of the matrix
                  a[i,k+1:n] = a[i,k+1:n] - lam*a[k,k+1:n]
                  #we update vector b
                  b[i] = b[i] - lam*b[k]
                  # backward substitution
      for k in range(n-1,-1,-1):
          b[k] = (b[k] - np.dot(a[k,k+1:n],b[k+1:n]))/a[k,k]

      return b

  aOrig = A.copy() # save original matrix A
  bOrig = B.copy() #save original vector b
  X = gauss_metod(aOrig, bOrig)
  #print A transformed for check
  print('n =', n)
  print('Метод Гаусса:')
  print("X =\n",X)
  #print("\nCheck result: [a]{x} - b =\n",np.dot(A,x) - B)
  R = B - np.matmul(A,X)
  print("R =\n",R)

  def zeidel_metod(A, b, eps):
      n = len(A)
      x = np.zeros(n)  # zero vector

      converge = False
      while not converge:
          x_new = np.copy(x)
          for i in range(n):
              s1 = math.fsum(A[i][j] * x_new[j] for j in range(i))
              s2 = math.fsum(A[i][j] * x[j] for j in range(i + 1, n))
              x_new[i] = (b[i] - s1 - s2) / A[i][i]

          converge = np.sqrt(math.fsum((x_new[i] - x[i]) ** 2 for i in range(n))) <= eps
          x = x_new

      return x
  X1 = zeidel_metod(A, Btransp, 1e-5)
  X1transp = np.ones((n, 1))
  for i in range (0, n):
    X1transp[i][0] = X1[i]
  print('Метод Зейделя:')
  print("X =\n",X1transp)
  #print("\nCheck result: [a]{x} - b =\n",np.dot(A,x) - B)
  R = B - np.matmul(A,X1transp)
  print("R =\n",R)
```
